# Data Cleaning & EDA

##Populating the null values or giving some meaningful names to the data in different columns.

```python
new_terror['City'].fillna('Unknown', inplace=True)

new_terror['State'].fillna('Unknown', inplace=True)

new_terror['Target_Subcategory'].fillna('Unknown', inplace=True)

new_terror['Weapon_subtype'].fillna('Unknown', inplace=True)

new_terror['Property_destroyed'].fillna('Undisclosed', inplace=True)

new_terror['Latitude'].fillna(0, inplace=True)

new_terror['Longitude'].fillna(0, inplace=True)

new_terror['Count_of_terrorist_involved'].fillna(0, inplace=True)

new_terror['Peoples_killed'].fillna(0, inplace=True)

new_terror['US_citizens_killed'].fillna(0, inplace=True)

new_terror['Terrorist_killed'].fillna(0, inplace=True)

new_terror['Peoples_injured'].fillna(0, inplace=True)

new_terror['Kids_hostage'].fillna(0, inplace=True)

new_terror['Ransom_demanded'].fillna(0, inplace=True)
```

```python
new_terror['Value_of_property_destroyed'] = new_terror['Value_of_property_destroyed'].fillna(new_terror
                                            ['Value_of_property_destroyed'].mode()[0])
```

```python
new_terror['Total_injured/deceased'] = new_terror['Peoples_killed'] + new_terror['Terrorist_killed'] + new_terror['Peoples_injured']
```

# Data visualization

```python
plt.subplots(figsize=(13,6))
sns.barplot(terror['Year'].value_counts().index, terror['Year'].value_counts().values)
plt.xticks(rotation=90)
plt.title('Terrorist Attacks By Year')
plt.show()
```

```python
plt.subplots(figsize=(13,6))
sns.countplot('Type_Of_Attack', data =terror, palette='inferno', order=terror['Type_Of_Attack'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Category Of Attacks')
plt.show()
```

```python
plt.subplots(figsize=(13,6))
sns.countplot(terror['Target_Category'], palette='Set3',order=terror['Target_Category'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Top Target Category')
plt.show()
```

```python
tg=terror['Terrorist_Group'].value_counts()
```

```python
plt.subplots(figsize=(15,6))
sns.barplot(tg[1:11].values, tg[1:11].index, palette='mako')
plt.xticks(rotation=0)
plt.xlabel('Count Of Attacks')
plt.title('Top 10 Terrorist Organizations',fontsize='20')
plt.show()
```
